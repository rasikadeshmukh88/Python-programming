## Example of Numeric data type

```python
x = 10
print(type(x))  # Output: <class 'int'>
```

```python
y = 3.14
print(type(y))  # Output: <class 'float'>
```

```python
z = 2 + 3j
print(type(z))  # Output: <class 'complex'>
```

## Example of String data type

```python
text = "Hello, world!"
print(type(text))  # Output: <class 'str'>
```

## Example of Boolean data type(True and False)

```python
is_true = True
print(type(is_true))  # Output: <class 'bool'> # bool type includes True and False
```

## Example of None Data Type

```python
x = None
print(type(x))  # Output: <class 'NoneType'>
```

# Python Type Casting Examples

Type casting in Python means converting one data type into another.

```python
x = 2.3
y = int(x)
print(y)
print(type(x))
print(type(y))
```

```python
#  int() ‚Äì Convert to integer
x = int(3.7)      # 3
y = int("10")     # 10

print(x)
print(y)
print(type(x))
print(type(y))
```

```python
# float() ‚Äì Convert to float
x = float(5)      # 5.0
y = float("2.3")  # 2.3
z = float('2')
print(z)
print(type(z))
print(x)
print(y)
print(type(x))
print(type(y))
```

```python
# str() ‚Äì Convert to string
x = str(25)       # "25"
y = str(3.14)     # "3.14"
print(x)
print(y)
type(x)
```

```python
# bool() ‚Äì Convert to boolean
bool(0)           # False
# bool(1)           # True
bool("")          # False
# bool("Hello")   # True
bool(" ")     # True
```

```python
# int(), float(), string() type conversion to complex

x= complex(5)           # (5+0j)
y =complex(2.5)         # (2.5+0j)
z =complex("3+4j")      # (3+4j)
type(z)
```

Note üòé : Non-numeric strings cannot be type cast to int, float, or complex in Python.
Attempting to do so will raise a ValueError.
