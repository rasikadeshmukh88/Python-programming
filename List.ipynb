```python
#List methods
#A)Adding elements in the list:append(),insert(),extend()
#1)append():only one element can be added at the end of list
name=["Rasika","shruti","Pranali"]
print("List Before adding element:",name)
name.append("Sakshi")
print("List after adding element:",name)
```

```python
#addes list in the list using append()
name=["Rasika","shruti","Pranali"]
print("List Before adding element:",name)
name.append(["Sakshi","payal"])
print("List after adding element:",name)
```

```python
marks = [67, 98, 100, 56]
print("List before adding elements",marks)
marks.append(89)
marks.append([76,34])
print("List after adding elements",marks)
```

```python
#2)insert():element can be added at desired position it takes two arguments
fruits = ['apple', 'mango','pineapple']
print(fruits)
fruits.insert(2, 'guava')
fruits.insert(1, ['lemon', 'lichi'])
print(fruits)
```

```python
#extend():added multiple elements in the list at the end of list
name=["Rasika","shruti","Pranali"]
name.extend(["sakshi","payal","Nikita"])
print(name)
```

```python
#B)Removing elements from the list:remove(),pop()
#1)remove():remove one element at a time(first occurrence)
fruits = ['apple', 'mango', 'pineapple','apple', 'guava']
fruits.remove('mango')
print(fruits)
```

```python
#2)pop():removes by default last element and returns the removed element
#to remove specific  element pass an argument(position)
fruits = ['apple', 'mango', 'pineapple','apple', 'guava']
print(fruits.pop())  #returned last element of the list
print(fruits.pop(2))  #returned element at 2 index
```

```python
#Reversing a List
fruits = ['apple', 'mango', 'pineapple','apple', 'guava']
fruits.reverse()      #using reverse() method
print(fruits)

#s=fruits[::-1]       #using slicing
#print(s)
len(fruits) # give the length of the list
```

```python
fruits = ['apple', 'mango', 'pineapple','apple', 'guava']
fruits.clear()  #it will clear the list
print(fruits)
```

```python
# sort() :sort in ascending order
num = [83,98, 3, 7, 54, 35, 8]
num.sort()
print(num)
```

```python
# sort() :sort in descending order
num = [83,98, 3, 7, 54, 35, 8]
num.sort(reverse = True)
print(num)
```

```python
#mixed data values sorting is followed by ASCII values for sorting
mixed_list = ['vadapav', 'Dosa', 'alootikki', 'Cholebhature', 'chaat', '#', '%']
mixed_list.sort()
print(mixed_list)
```

```python
mixed_list = ['vadapav', 'Dosa', 'alootikki', 'Cholebhature', 'chaat', '#', '%']
mixed_list.sort(reverse=True)
print(mixed_list)
```
