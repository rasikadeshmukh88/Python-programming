```python
# Base class
class Person:
    def __init__(self, name, age):
        self.name = name          # Public attribute
        self.__age = age          # Private attribute (encapsulation)

    # Public method to get private age
    def get_age(self):
        return self.__age

    # Public method to set private age
    def set_age(self, new_age):
        if new_age > 0:
            self.__age = new_age
        else:
            print("Age must be positive.")

    def display_person_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.__age}")

# Derived class (inherits from Person)
class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)         # Call base class constructor
        self.student_id = student_id        # Public attribute
        self.__grade = grade                # Private attribute

    # Public method to access private grade
    def get_grade(self):
        return self.__grade

    # Public method to modify private grade
    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("Grade must be between 0 and 100.")

    def display_student_info(self):
        self.display_person_info()   # Call base class method
        print(f"Student ID: {self.student_id}")
        print(f"Grade: {self.__grade}")

# Example usage
student1 = Student("Rasika", 20, "S1455", 88)

# Access public attribute
print("Student Name:", student1.name)

# Access private attribute via public method
print("Initial Grade:", student1.get_grade())
student1.set_grade(92)

# Try to set invalid grade
student1.set_grade(105)

# Access and modify private age via method
print("Age:", student1.get_age())
student1.set_age(21)

print("\nFull Student Info:")
student1.display_student_info()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age

class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)
        self.student_id = student_id
        self.__grade = grade

    def get_grade(self):
        return self.__grade

    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade

    def display(self):
        print(f"Name: {self.name}, Age: {self.get_age()}, ID: {self.student_id}, Grade: {self.__grade}")

# Example usage
s = Student("Rasika", 20, "S1455", 85)
s.set_grade(90)
s.set_age(21)
s.display()
```
