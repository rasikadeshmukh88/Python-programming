# Sets

Sets in Python used to store a collection of items with the following properties.

* No duplicate elements.
* An unordered collection. When we access all items, they are accessed without any specific order and we cannot access items using indexes as we do in lists.
Internally use hashing that makes set efficient for search, insert and delete operations.
* Mutable, meaning we can add or remove elements after their creation, the individual elements within the set cannot be changed directly.

## Check unique and  Immutable with Python Set

### Adding and removing elements from the set

Adding and Removing elements
We can add and remove elements form the set with the help of the below functions â€“

* add(): Adds a given element to a set
* clear(): Removes all elements from the set
* discard(): Removes the element from the set
* pop(): Returns and removes a random element from the set
* remove(): Removes the specified element from the set
                 
if element is missing in remove()-- it Raises KeyError (error) but in discard--it Does nothing (no error)


```python
# set of letters
s = {'g', 'e', 'k', 's'} 

# adding 'f'
s.add('f')
print('Set after updating:', s)
```

```python
s = {'g', 'e', 'k', 's'}
# Discarding element from the set
s.discard('e')
print('Set after updating:', s)
```

```python
s = {'g', 'e', 'k', 's'}
# Removing element from the set
s.remove('f')
print('\nSet after updating:', s)
```

```python
s = {'g', 'e', 'k', 's'}
# Popping elements from the set
print(s.pop())
# print('Set after updating:', s)
```

```python
s.clear()
print('\nSet after updating:', s)
```

## update()

The update() method for sets is used to add elements from another set, list, tuple, or any iterable to the set

syntax ---- > set.update(*others)

it help to add multiple elements
*others: This can be one or more iterables (like sets, lists, tuples, etc.).
Duplicate elements are ignored because sets only store unique elements.

```python
#example when we are adding element using add

# sample = {2,3,5}
# sample.add([3,4])
# print(sample)      #TypeError: unhashable type: 'list'

sample = {2,3,5}
sample.add((3,4))
print(sample)   # since tuple is immutable so it has added but as one element in same index
```

```python
# example of adding element using update()
sample = {2,3,5}
sample.update([8,9])
print(sample)   # by using update we have added different elements
```

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1.update(set2)
print(set1)
```

## union()

* Union of A and B is a set of all elements from both sets
* also denoted by the operator |

```python
A = {1, 2, 3, 4, 5}
B= {4, 5, 6 , 7 , 8}

A.union(B)
```

## intersection()

* intersection() method returns the intersection of two sets
* Also denoted by the operator &
* Intersection of A and B is a set of elements that are common in both the sets.

```python
A = {1, 2, 3, 4, 5}
B= {4, 5, 6 , 7 , 8}
```

```python
print(A & B)
print(B & A)
```

```python
print(A.intersection(B))
```

```python
# printing intersection without using .intersection operator
print(B|A)
print(A^B)
print((B|A) - (A^B))
```

## Difference
* difference() method returns the difference of two sets
* Also denoted by the operator -
* Return a new set with elements in the set that are not in the others.

```python
A = {1, 2, 3, 4, 5}
B= {4, 5, 6 , 7 , 8}
```

```python
#A - B
print(A.difference(B)) #A.difference(B) returns a new set containing elements that are in A but not in B.

#B-A
print(B.difference(A))
```

## Symmetric_difference()
* denoted by the operator ^
* symmetric_difference() method returns all the elements without the intersection
* The symmetric difference between two sets is the set of all the elements that are either in the first set or the second set but not in both.

```python
A = {1, 2, 3, 4, 5}
B= {4, 5, 6 , 7 , 8}
```

```python
print(A ^ B)
print(A.symmetric_difference(B))
```

# Frozen Sets

* A frozenset is like a set, but it's immutable (canâ€™t be changed after creation).

* Useful when you need a set as a dictionary key or want to ensure the set stays constant.

Syntax : frozenset(iterable_object_name)

```python
# Python program to demonstrate differences
# between normal and frozen set

# Same as {"a", "b","c"}
s = set(["a", "b","c"])

print("Normal Set")
print(s)

# A frozen set
fs = frozenset(["e", "f", "g"])

print("\nFrozen Set")
print(fs)

# Uncommenting below line would cause error as
# we are trying to add element to a frozen set
fs.add("h")
```

```python
animals = frozenset(["cat", "dog", "lion"])
print("cat" in animals)
print("elephant" in animals)
```

```python
# initialize A and B
A = frozenset([1, 2, 3, 4])
B = frozenset([3, 4, 5, 6])

# copying a frozenset
C = A.copy()
print(C)

# union
union_set = A.union(B)
print(union_set)

# intersection
intersection_set = A.intersection(B)
print(intersection_set)

difference_set = A.difference(B)
print(difference_set)

# symmetric_difference
symmetric_difference_set = A.symmetric_difference(B)
print(symmetric_difference_set)
```
