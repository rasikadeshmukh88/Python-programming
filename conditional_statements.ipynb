```python
#Example of if-else statement
a=int(input("Enter first number a="))
b=int(input("Enter second number b="))
if a<b:
    print("a is less than b")
else:
    print("a is not less than b")
```

```python
text="HELLO"
if "s" in text:
    print("s is present in the text")
else:
    print("s is not present in the text")
```

```python
num=int(input("Enter the number :"))
if num % 2==0:
    print("Number is even")
else:
    print("Number is odd")
```

```python
#Examples of elif statements
num=int(input("Enter the number :"))
if num>0:
    print("Number is greater than zero")
elif num<0:
    print("Number is less than zero")
else:
    print("this number is zero")
```

```python
marks=int(input("Enter the student marks between()1-100:"))
if marks>90:
    print("grade:A+")
elif marks>=80 and marks<90:
    print("grade:B")
elif marks>=70 and marks<80:
    print("grade:C")
elif marks>=60 and marks<70:
    print("grade:D")
else:
    print("Fail")
```

```python
#Example of nested if amd if-else statement
num=int(input("Enter the number:"))
if num>0:
    if num % 2==0:
        print("positive even number")
    else:
        print("positive odd number")
else:
    print("number is less than zero")
```

```python
#shorthand ef-else statement
x=int(input("Enter the number :"))
print("Even Number") if x%2==0 else print("Odd Number")
```
