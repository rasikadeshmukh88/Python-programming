```python
#Arithmetic Operators
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
print("Addition :",a+b)
print("Subtraction :",a-b)
print("Multiplication :",a*b)
print("Division :",a/b)
print("Modulus :",a%b)
print("power :",a**b)
print("Floor :",a//b)
```

```python
#Comparison operator
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
print("a equals to b:",a==b)
print("a not equals to b:",a!=b)
print("a greater than b:",a>b)
print("a less than b:",a<b)
print("a greater than or equals to b:",a>=b)
print("a less than or equals to b:",a<=b)
```

```python
#Bitwise Operators
a = 5     # 0101
b = 3     # 0011
print(a & b)  # 1 (0001)
print(a | b)  # 7 (0111)
print(a ^ b)  # 6 (0110)
print(a << 1) # 10 (1010)
print(a >> 1) # 2 (0010)
```

```python
a = 8
print(a<<2)
a = 8
print("Original a     :", bin(a))       # binary of 8
print("a << 2         :", bin(a << 2))  # binary after left shift
print("Decimal result :", a << 2)       # decimal value
```

```python
# Identity Operators
x = [1, 2]
y = [1, 2]
z = x
print(x is z)       # True (same object)
print(x is y)       # False (different objects)
print(x == y)       # True (same content)
```

```python
#Membership operators
my_list = [1, 2, 3, 4]
print(2 in my_list)     # True
print(1 not in my_list) # False
```

```python
# difference between is and == operator
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)   # True → values are equal
print(a is b)   # False → different objects in memory
print(a is c)   # True → same object
```
