#### Example of zip elements from list

```python
names=["Rasika","pranali","Shruti"]
age=[21,22,23]
zip(names,age)

info=list(zip(names,age))
print(info)

#2. Unzip elements from the list of tuples

names, age = zip(*info)
print("names:",names)
print("age:",age)
```

#### 4. Write a Python program that has a predefined list of fruits: ["apple", "banana", "mango", "grape"].Takes a fruit name as input from the user.
#### Checks whether the entered fruit is present in the list.  If found, adds it to a new empty list and prints the updated list. 
#### If not found, prints "Not found in the list".

```python
fruits=["apple","banana","mango","grape"] #predefined list
print("List of fruits:",fruits)

update_list=[]                             #empty list for store found fruits
fruit_name=input("Enter fruit name:")      #takes user input

if fruit_name in fruits:
    print("fruit is found in the present list")
    update_list.append(fruit_name)
    print("updated list for found fruits:",update_list)   #print update list
    
else:
    print("Not found in the list")
```

#### 5. Write a program to insert an element in the tuple at a specificÂ position.

```python
colors=("Red","Pink","Black","Green")
print(colors)   #printing tuple

x=list(colors)  #converting tuple into list
print(x)        #printing list

x.insert(2,"Yellow")  #insert "yellow" at index 2
print(x)

colors=tuple(x)  #converting list into tuple
print(colors)
```
