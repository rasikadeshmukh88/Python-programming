```python
### Write a Program to print numbers
n=1
row=4
for  i in range(1,row+1):
    for j in range(i):
        print(n,end=" ")
        n+=1
    print()
```

```python
#Write a program to skip weekends in a work scheduler
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
for day in days:
    if day in ["Sat", "Sun"]:
        continue
    print("Working on", day)
```
