## String Slicing Examples

```python
# # String slicing examples
text = "Electronics"
print(text)
# # Basic slices
print("Original text:", text)
print("text[0:5]     :", text[0:5])      # Output: Elect
print("text[2:9:2]   :", text[2:9:2])    # Output: etoi
print("text[-3:]     :", text[-3:])      # Output: ics
print("text[:4]      :", text[:4])       # Output: Elec
print("text[::2]     :", text[::2])      # Output: Eetoinc
```

```python
# program to reverse a string
text = "datascience"
print(text[::-1])
```

# join() - we use join() method to combine a list of strings into a single string with each string separated by a space

#### Syntax :
    separator.join(iterable)
    
###### Parameters:
* separator: The string placed between elements of the iterable.
* iterable: A sequence of strings (e.g., list, tuple etc) to join together.

```python
# # joining a string
# Syntax - separator.join(iterable)
words = ["Python", "is", "fun"]
sentence = " ".join(words)
print(sentence)  # Output: Python is fun
```

```python
fruits = ['apple', 'banana', 'orange']
result = "-".join(fruits)
print(result)
```

# split() method splits a string into a list of strings after breaking the given string by the specified separator.

##### Syntax:
    str.split(separator, maxsplit)
    
##### Parameters
* separator: This is a delimiter. The string splits at this specified separator. If is not provided then any white space is a separator.
* maxsplit: It is a number, that tells us to split the string into a maximum of the provided number of times. If it is not provided then the default is -1 which means there is no limit.

```python
# # Splitting a string
text = "apple,banana, grape"
fruits = text.split(",")
print(fruits)  # Output: ['apple', 'banana', 'grape']
```

```python
data = "name-age-city"
parts = data.split("-")
print(parts)
```

```python
data = "name-age-city"
parts = data.split("-", maxsplit = 1) #The string is split at each : and returned as a list of parts.
print(parts)
```

# strip() method in Python removes all leading and trailing whitespace by default.You can also specify a set of characters to remove from both ends. It returns a new string and does not modify the original.

##### Syntax :
    s.strip(chars)
    
##### Parameters:

* chars(optional): A string specifying the set of characters to remove from the beginning and end of the string.
* Return Type: A new string with specified characters removed from both ends.

```python
# # strip()
text = "  hello  "
print(text)
print(text.strip())  # Output: "hello"
```

```python
# strip()
url = "www.example.com.www"
cleaned = url.strip("w.")  # Removes 'w' and '.' from both ends
print(cleaned)
```
