# Day 15 ---  Builtin Functions, Date Time and  Modules

```python
# Sample data
numbers = [10, -4, 7, 0, 15]

# SUM
total = sum(numbers)
print("SUM:", total)

# AVG (mean) - Python has no direct avg built-in, use sum/len or statistics.mean
average = sum(numbers) / len(numbers)
print("AVG:", average)

# ABS
abs_values = [abs(n) for n in numbers]
print("ABS (each number):", abs_values)
print("ABS of -4:", abs(-4))

# MIN and MAX
print("MIN:", min(numbers))
print("MAX:", max(numbers))
```

# Date Time

Python Datetime module comes built into Python, so there is no need to install it externally.

```python
# imported date and time from datetime module
from datetime import date, time

# want to print today's date, month and year

today = date.today()
today
```

```python
# want to print date of today
day = today.day
day
```

```python
# want to print the month
month = today.month
month
```

```python
# want to print the year
year = today.year
year
```

```python
from datetime import datetime
now = datetime.now()
print("Current date and time: ", now)

print("Date only :", now.date())
print("Time only: ", now.time())
```

```python
# want to print in string format
print(today.strftime("%A %d %B %Y"))
```

```python
print(today.strftime("%A, %dth of %B %Y"))
```

```python
print(today.strftime("%a %d %b %y"))
```

```python
# want to know the next year
next_year = today.replace(year = today.year + 1)
print(next_year)
```

```python
next_year_month = today.replace(year = today.year + 1, month = today.month + 1)
print(next_year_month)
```

```python
# some calculation difference from today
diff = abs(next_year - today)
print(f' Only {diff.days} days untill next year')
```

```python
apjabdulkalam = date(1931, 10, 15)
# which day of the week was it?
apjabdulkalam.weekday()  # 0-- sunday, 1-- monday, 2--- tuesday, 3 --- wednesday....
```
