```python
#names = ['IN001', 'US002', 'PK005','IN002','IN005','IN003'] Write a function that takes a list of codes 
#and separates them into different lists based on whether they start with "IN", "PK", or "US", then prints each list.
names = ['IN001', 'US002', 'PK005', 'IN002', 'IN005', 'IN003']
def separate_codes(codes):
    in_codes = []
    pk_codes = []
    us_codes = []

    for code in codes:
        if code.startswith("IN"):
            in_codes.append(code)
        elif code.startswith("PK"):
            pk_codes.append(code)
        elif code.startswith("US"):
            us_codes.append(code)

    print("IN codes:", in_codes)
    print("PK codes:", pk_codes)
    print("US codes:", us_codes)

separate_codes(names)
```

```python
#Write a function that counts the number of vowels in a string.
count_vowels = lambda s: len(list(filter(lambda ch: ch.lower() in 'aeiou', s)))
print(count_vowels("congratulations"))
```

```python
#Write a recursive function to find the power of a number (e.g., 2^5).
def power(base,exponent):
    if exponent==0:
        return 1
    else:
        return base*power(base,exponent-1)

print("2^5 :",power(2,5))
```

```python
# numbers = [2, 5, 8, 10, 15, 21, 30]
# Use a lambda function with filter() to extract all numbers that are divisible by 5.
numbers=[2,5,8,10,15,21,30]
num=filter(lambda x:x%5==0,numbers)
print(list(num))

# Use a lambda function with map() to square each number in the list.
numbers=[2,5,8,10,15,21,30]
square=map(lambda x:x**2,numbers)
print(list(square))

# Use a lambda function with reduce() to calculate the product of all numbers in the list.
from functools import reduce
numbers=[2,5,8,10,15,21,30]
product=reduce(lambda x,y:x*y,numbers)
print(product)

#  Expected output
# Divisible by 5: [5, 10, 15, 30]
# Squared List: [4, 25, 64, 100, 225, 441, 900]
# Product of all numbers: 75600000
```
