```python
#1)Write a function count_vowels(s) that returns the number of vowels in the given string.
vowels="aeiou"
def count_vowels(s):
    count=0
    for i in s:
        if i in vowels:              
            count+=1
    return count
    
s=input("Enter a string:")
print("Number of vowels:",count_vowels(s))

count_vowels(s) 
```

```python
#2)Write a function calculator(a, b, operation) where operations can be 'add', 'subtract', 'multiply', 'divide'.
def calculator(a,b,operation):
    if operation=="add":
        return a+b
    elif operation=="subtract":
        return a-b
    elif operation=="multiply":
        return a*b
    elif operation=="divide":
        return a/b
    else:
        print("Enter valid operation")

a=float(input("Enter value for a:"))
b=float(input("Enter value for b:"))
operation=input("Enter operation (add,subtract,multiply,divide:)")

calculator(a,b,operation)
```
