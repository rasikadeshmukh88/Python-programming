```python
#3.Write a program to take user input and then find whether a number is prime number or not.
num=int(input("Enter the number:"))
if num<=1:
    print("Not prime number")
else:
    for i in range(2,num):
        if num%i==0:
            print("Not prime number")
            break
        else:
            print("prime number")
            break
```

```python
#check the number is prime or not
num=int(input("Enter the number"))
for i in range(2,(num**0.5)+1):
    if num%i==0:
        print("not prime number")
    else:
        print("prime number")
```
