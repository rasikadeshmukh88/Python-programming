```python
#Type casting:Converting one data type into another
x=2.3   #float
y=int(x)  #convert into integer
print(y)
print(type(x))
print(type(y))
```

```python
#int()-convert to integer
x=int(3.7)    # 3 float to int
y=int("10")   #10 string to int
print(x)
print(y)
print(type(x))
print(type(y))
```

```python
#float()-convert to float
x=float(5)   #5.0
y=float("2.5")    #2.5
z=float('2')
print(z)    #2.0
print(type(z))
print(x)
print(y)
print(type(x))
print(type(y))
```

```python
#str()-convert to string
x=str(25)
y=str(3.14)
print(x)
print(y)
print(type(x))
print(type(y))
```

```python
#bool()-convert to boolean
bool(0)
print(bool(0))      # False

bool(1)
print(bool(1))      #True

print(bool(""))      #False
print(bool("Hello"))  #True
print(bool(" "))      #True
```

```python
#int(),float(),string() type conversion to complex
x=complex(5)
print(x)

y=complex(2.5)
print(y)

z=complex("3+4j")
print(z)

print(type(x))
print(type(y))
print(type(z))
```
