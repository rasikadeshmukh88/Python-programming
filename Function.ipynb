## yield
In Python, yield is used in a function to turn it into a generator. Instead of returning a single value and ending the function (like return), yield pauses the function and saves its state. The function can later be resumed from where it left off.

```python
def even_generator(limit):
    for i in range(2, limit + 1, 2):
        return i

print(even_generator(10))
```

```python
def even_generator(limit):
    for i in range(2, limit + 1, 2):
        yield i

#print(even_generator(10))
for num in even_generator(10):
    print(num)
```

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for number in count_up_to(3):
    print(number)
```

# Recursive Function
* A recursive function in Python is a function that calls itself to solve a smaller version of a problem until
it reaches a base case (a condition where it stops calling itself).
* Often used to solve the problems that ca be broken into smaller, self-similar subproblems.

```python
def factorial(n):
    if n == 0:
        return 1  # Base case
    else:
        return n * factorial(n-1)  # Recursive call

# get the user input
num = int(input("Enter a non negative integer "))
result = factorial(num)
print(f'The Factorial of {num} is', result)
```

```python
factorial ((6)) # Here, factorial(5) calls factorial(4), then factorial(3), and so on, until it reaches factorial(0).
```

```python
# Sum of first n natural numbers
def sum_natural(n):
    if n == 1:
        return 1  # Base case
    else:
        return n + sum_natural(n-1)  # Recursive call sum_natural(5) ‚Üí 5 + 4 + 3 + 2 + 1 = 15
```

```python
sum_natural(9)
```
